```python
import pandas as pd
data={
     "Country": ["Pakistan", "India", "China", "USA", "UK"],
     "Cases": [15000, 25000, 8000, 40000, 12000],
     "Recovered": [14000, 23000, 7800, 39000, 11500],
     "Deaths": [200, 500, 50, 800, 150]
}

df=pd.DataFrame(data)
df
```

```python
# Total Cases
print(df["Cases"].sum())
```

```python
# Average Cases
print(df["Cases"].mean())
```

```python
# Highest Cases
print(df["Cases"].max())
```

```python
print(df[df["Cases"] == df["Cases"].max()])
```

```python
# Recovery Rate
df["Recovery Rate"]=(df["Recovered"]/df["Cases"])*100
df
```

```python
df["Death Rate"]=(df["Deaths"]/df["Cases"])*100
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x="Country",y="Cases",data=df)
plt.xlabel("Country",color="red",size=15)
plt.ylabel("Cases",color="red",size=15)
plt.title("COVID Cases",color="blue",size=20)
plt.show()
```

```python
plt.pie(df["Cases"],labels=df["Country"], autopct="%1.1f%%")
plt.title("COVID Cases Distribution")
plt.show()
```
