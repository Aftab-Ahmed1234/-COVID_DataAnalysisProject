```python
import pandas as pd
data={
     "Country": ["Pakistan", "India", "China", "USA", "UK"],
     "Cases": [15000, 25000, 8000, 40000, 12000],
     "Recovered": [14000, 23000, 7800, 39000, 11500],
     "Deaths": [200, 500, 50, 800, 150]
}

df=pd.DataFrame(data)
df
```

```python
# Total Cases
print(df["Cases"].sum())
```

```python
# Average Cases
print(df["Cases"].mean())
```
